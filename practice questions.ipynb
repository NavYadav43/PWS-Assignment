```python

# File path: string_list_tuple_set_problems_solutions.py

from collections import Counter
import re
from itertools import islice, accumulate
import random
import math


# STRING-BASED PROBLEMS
def reverse_string(s):
    return s[::-1]

def is_palindrome(s):
    return s == s[::-1]

def to_uppercase(s):
    return s.upper()

def to_lowercase(s):
    return s.lower()

def count_vowels(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)

def count_consonants(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char.isalpha() and char not in vowels)

def remove_whitespace(s):
    return s.replace(' ', '')

def string_length(s):
    count = 0
    for _ in s:
        count += 1
    return count

def contains_word(s, word):
    return word in s

def replace_word(s, old, new):
    return s.replace(old, new)

def count_word_occurrences(s, word):
    return s.split().count(word)

def find_first_occurrence(s, word):
    return s.find(word)

def find_last_occurrence(s, word):
    return s.rfind(word)

def split_string(s):
    return s.split()

def join_words(words):
    return ' '.join(words)


# LIST-BASED PROBLEMS
def create_list(start, end):
    return list(range(start, end + 1))

def list_length(lst):
    count = 0
    for _ in lst:
        count += 1
    return count

def append_element(lst, elem):
    lst.append(elem)
    return lst

def insert_element(lst, index, elem):
    lst.insert(index, elem)
    return lst

def remove_element_by_value(lst, value):
    lst.remove(value)
    return lst

def remove_element_by_index(lst, index):
    del lst[index]
    return lst

def element_exists(lst, elem):
    return elem in lst

def index_of_element(lst, elem):
    return lst.index(elem)

def count_element_occurrences(lst, elem):
    return lst.count(elem)

def reverse_list(lst):
    return lst[::-1]


# TUPLE-BASED PROBLEMS
def create_tuple(start, end):
    return tuple(range(start, end + 1))

def access_tuple_element(tpl, index):
    return tpl[index]

def tuple_length(tpl):
    count = 0
    for _ in tpl:
        count += 1
    return count

def count_tuple_element(tpl, elem):
    return tpl.count(elem)

def index_of_tuple_element(tpl, elem):
    return tpl.index(elem)


# SET-BASED PROBLEMS
def create_set(start, end):
    return set(range(start, end + 1))

def add_to_set(s, elem):
    s.add(elem)
    return s

def remove_from_set(s, elem):
    s.remove(elem)
    return s

def union_of_sets(s1, s2):
    return s1.union(s2)

def intersection_of_sets(s1, s2):
    return s1.intersection(s2)

def difference_of_sets(s1, s2):
    return s1.difference(s2)

```
